```python
# figure 명시해줘야 할 필요성 있음 (사용할 때마다)
plt.figure()
plt.plot(1, 1, 'o')
plt.plot(2, 2, 'o')
plt.plot(3, 3, 'o')
```

```python
plt.figure()
plt.plot(1, 1, 'o')
plt.plot(2, 2, 'o')
plt.plot(3, 3, 'o')
# scale 변경
ax = plt.gca()
ax.axis([0, 5, 0, 10])
```

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
# x = [1, 2, 3, 4, 5, 6, 7, 8]
y = x

plt.figure()
plt.plot(x, y, 'og')
```

```python
colors = ['green'] * (len(x) - 1)
colors.append('red')

plt.figure()
plt.scatter(x, y, c=colors)
```

```python
plt.figure()
plt.scatter(x[:3], y[:3], c='red', label='Small')
plt.scatter(x[3:], y[3:], c='gray', label='Large')
plt.xlabel('Number of success')
plt.ylabel('Grades')
plt.title('Relationship between success and grade')
# legend: 범례
plt.legend()
```

# Line Plots

```python
linear_data = x
linear_data
```

```python
exponential_data = linear_data**2

plt.figure()
plt.plot(x, linear_data, '-.o', x, exponential_data, '--o')
```

```python
plt.figure()
plt.plot(x, linear_data, '-o', x, exponential_data, '-o')
plt.plot(x[:3], [22, 44, 55], '--r')
plt.legend(['Baseline', 'Completion', 'Test'])
```

```python
plt.figure()
plt.plot(x, linear_data, '-o', x, exponential_data, '-o')
plt.plot(x[:3], [22, 44, 55], '--r')
plt.legend(['Baseline', 'Completion', 'Test'])
plt.fill_between(x, linear_data, exponential_data, facecolor='gray', alpha=0.2)
```

# Bar Charts

```python
plt.figure()
plt.bar(x, linear_data, width=0.5)
```

```python
plt.figure()
plt.bar(x, linear_data, width=0.3)
plt.bar(x+0.3, exponential_data, width=0.3)
```

```python
plt.figure()
plt.bar(x, linear_data, width=0.3)
plt.bar(x, exponential_data, bottom=linear_data, width=0.3, color='gray')
```
