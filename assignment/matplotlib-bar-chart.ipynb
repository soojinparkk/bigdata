```python
import matplotlib.pyplot as plt
import numpy as np
```

1. 밑쪽의 라벨을 제외하고 모든 라벨과 tick들과 테두리들을 삭제   
2. Python을 제외한 나머지 막대들을 회색으로 바꾸고, Python 은 파랑색으로 유지   
3. 크기에 해당하는 라벨을 그래프 안에 써주는데 그 x위치는 막대x에 너비를 2로 나눈 것을 더하고, y위치는 막대높이 빼기 5만큼 위치를 설정

```python
plt.figure()
languages = ['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]
x = plt.bar(pos, popularity, align='center')
plt.xticks(pos, languages)
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data', alpha=0.8)
```

```python
# 1-1. 모든 테두리 삭제
ax = plt.gca()
loc = ['left', 'right', 'top', 'bottom']
for rm in loc:
    ax.spines[rm].set_visible(False)
# 1-2. x축 label을 제외하고 tick만 삭제
plt.tick_params(axis='x', bottom=False)
# 1-3. y축 tick 삭제
ax.set_yticks([])
# 1-4. y축 label 삭제
ax.axes.yaxis.set_visible(False)

# 2. python을 제외한 나머지 막대 회색으로 설정
for i in pos:
    if (i == 0):
        continue
    x[i].set_color('gray')

# 3. 바 크기 표시 및 위치 설정
for p in x.patches:
    left, bottom, width, height = p.get_bbox().bounds
    plt.annotate("%d%%"%(height), (left+width/2, height-5), ha='center', color='#FFFFFF', size=12)
```
