### 공공 데이터 포털 오픈 API
K-MOOC 강좌 조회 서비스

```python
import urllib.request
import json
```

```python
url = "http://apis.data.go.kr/B552881/kmooc/courseList?serviceKey=zUsap7I%2BSXPwfCt3zHKg0jwPW2gxxSTqSRvwqt0STgw8OwtHAUbkzsWo5azP3P0pv%2Bt9UoGBBEbfhthnMWU9dA%3D%3D&Page=1&Org=FUNMOOC&Mobile=1&type=json"

response = urllib.request.urlopen(url)
json_str = response.read().decode('utf-8')
```

```python
json_object = json.loads(json_str)
json_object
```

results 부분만 추출하여 dataframe 생성

```python
import pandas as pd

df = pd.DataFrame(json_object['results'])
df.head()
```

1. 강좌의 name을 index로 설정
2. 필요한 column만 추출

```python
df.set_index('name', inplace=True)
df = df.loc[:, ['short_description', 'teachers', 'level', 'passing_grade', 'org_name', 'classfy_name']]
df
```

passing_grade를 숫자로 형변환

```python
df['passing_grade'] = pd.to_numeric(df['passing_grade'])
```
