10000개의 sample을 가진 두개의 표준정규분포(평균 0, 표준편차 1)를 가지고 scatter plot (붉은색 점)을 그리고,

그 위와 오른쪽에 파란색의 normalized histogram을 그리세요.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 0:2])
side_histogram = plt.subplot(gspec[1:, 2])
lower_left = plt.subplot(gspec[1:, 0:2])
```

```python
x = np.random.normal(loc=0.0, scale=1.0, size=10000)
y = np.random.normal(loc=0.0, scale=1.0, size=10000)

lower_left.scatter(x, y, color='red')
t = top_histogram.hist(x, bins=100, density=True, color='blue')
s = side_histogram.hist(y, bins=100, orientation='horizontal', density=True, color='blue')
```
